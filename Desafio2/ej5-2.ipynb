```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
R = 1
N = 1000

theta = np.random.uniform(0,2*np.pi, N)
T = np.random.gamma(3,1, N)

X = R*T*np.cos(theta)
Y = R*T*np.sin(theta)

t = np.arange(0,N)
```

```python
fig, ax = plt.subplots(2, figsize=(10,10), sharex=True)

ax[0].set_title('X, Y')
ax[1].set_title('T')
ax[0].scatter(X, Y)
ax[1].hist(T)
ax[0].set(xlabel="X")
ax[0].set(ylabel="Y")
ax[1].set(xlabel="t")

plt.show()
```
